```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
import seaborn
```

```python
file_path = "/Users/aryellewright/Documents/Kumar-Biomaterials-Lab/CCK8_Assay_Plots/20231006_PolymerScreening_AW copy.xlsx" 
sheet_name = "Sheet2"  

# creating a dataframe from our xlsx file
df = pd.read_excel(file_path, sheet_name=sheet_name)
# print(df.head())
```

```python
# creating variables in our df to use to plot
variable = df['Variables'][::-1]
viability = df['Viability'][::-1]
std_devs = df['Standard Deviation'][::-1]
```

```python
# plotting the CCK8 assay information
plt.barh(variable, viability, xerr=std_devs, capsize = 3, color='gray')
plt.xlabel('Cell Viability (%)')
plt.show()
```

```python
# creating varibales from our df to plot for our transfection efficiency graph
transfection = df['Transfection'][::-1]
stddev = df['StdDev'][::-1]
```

```python
# plotting our GFP+ cells
plt.barh(variable, transfection, xerr=stddev, capsize = 3, color='gray')
plt.xlabel('% GFP +Ve Cells')
plt.show()
```
