```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "/Users/aryellewright/Documents/Documents - Aryelleâ€™s MacBook Air/Kumar-Biomaterials-Lab/CCK8_Transfection_Plots/20231211_CCK8_DIPRepeat(3).xlsx" 
sheet_name = "DIPPolyplexes"  

# creating a dataframe from our xlsx file
df = pd.read_excel(file_path, sheet_name=sheet_name)
# print(df.head())
```

```python
# creating variables in our df to use to plot
variable = df['Variables'][::-1]
viability = df['Viability'][::-1]
std_devs = df['Standard Deviation'][::-1]
```

```python
# plotting the CCK8 assay information
color_gray = "#818689"
plt.barh(variable, viability, xerr=std_devs, capsize = 3, color=color_gray)
plt.xlabel('Cell Viability (%)')
plt.show()
```

```python
# creating varibales from our df to plot for our transfection efficiency graph
transfection = df['Transfection'][::-1]
stddev = df['StdDev'][::-1]
```

```python
# plotting our GFP+ cells
plt.barh(variable, transfection, xerr=stddev, capsize = 3, color=color_gray)
plt.xlabel('% GFP +Ve Cells')
plt.xlim(left=0)

plt.show()
```
