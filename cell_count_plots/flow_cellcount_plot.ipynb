```python
import pandas as pd
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes
```

```python
file_path = "/Users/aryellewright/Documents/Kumar-Biomaterials-Lab/cell_count_plots/20231113_DIPRepeat_CellCount.xlsx" 
sheet_name = "Sheet1"  

# creating a dataframe from our xlsx file
df = pd.read_excel(file_path, sheet_name=sheet_name)
# print(df.head())
```

```python
# creating variables in our df to use to plot
variable = df['Variables'][::-1]
cell_count = df['Cell Count'][::-1]
std_devs = df['Standard Deviation'][::-1]
```

```python
# plotting the CCK8 assay information
plt.barh(variable, cell_count, xerr=std_devs, capsize = 3, color='gray')
plt.xlabel('Cell Count')
plt.show()
```

```python
# Create a broken axis plot
bax = brokenaxes(xlims=((0, 9000), (12000, 50000)))

# Plotting the CCK8 assay information
bax.barh(variable, cell_count, xerr=std_devs, capsize=3, color='gray')

# Setting labels
bax.set_xlabel('Cell Viability (%)')

# Show plot
plt.show()
```
