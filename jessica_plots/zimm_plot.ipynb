```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
df = pd.read_csv('/Users/aryellewright/Documents/Kumar-Biomaterials-Lab/jessica_plots/DIP50_HEMA50_zimm plot.csv', delimiter=',')
#print(df.head(2))
```

```python
# Given that each pair (sin^2(ϴ/2) + kc, Peak) represents one set of data to plot
columns = df.columns.tolist()
sets_of_data = len(columns) // 2  # Assuming pairs of columns

plt.figure(figsize=(12, 8))

# Plot each pair of columns
for i in range(sets_of_data):
    x_data = df[columns[2*i]]
    y_data = df[columns[2*i + 1]]
    if 'Fit' in columns[2*i+1] or 'Zero' in columns[2*i+1]:
        plt.plot(x_data, y_data, marker='', label=columns[2*i + 1])
    else:
        plt.plot(x_data, y_data, marker='o', linestyle='', label=columns[2*i + 1])

plt.title('Zimm Plot')
plt.xlabel('sin²(ϴ/2) + kc')
plt.ylabel('K*c/Rr$\dot{\Theta}$')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
