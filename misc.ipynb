```python
import numpy as np
import matplotlib.pyplot as plt

# Inverse temperatures (1/T)
inverse_temps = [1/298.15, 1/313.15]

# Natural log of CMC values
ln_CMCs = [np.log(0.92e-3), np.log(1e-3)]

# Plot the points
plt.plot(inverse_temps, ln_CMCs, 'o')

# Perform linear regression
slope, intercept = np.polyfit(inverse_temps, ln_CMCs, 1)

# Use the slope and intercept to plot the line of best fit
fit_line = np.poly1d([slope, intercept])
plt.plot(inverse_temps, fit_line(inverse_temps))

# Display the plot
plt.xlabel('1/T (1/K)')
plt.ylabel('ln(CMC)')
plt.show()

R = 8.314
# Outputs for enthalpy and entropy changes
Delta_H_mic_0 = -slope * R
Delta_S_mic_0 = intercept * R
```
